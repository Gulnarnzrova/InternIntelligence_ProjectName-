```python
import pandas as pd

file_path = r"C:\Users\User\Desktop\Chocolate Sales.csv"
df = pd.read_csv(file_path)

print(df.head())

print(df.dtypes)
```

```python
df["Amount"] = df["Amount"].replace("[$,]", "", regex=True).astype(float)

df["Date"] = pd.to_datetime(df["Date"], format="%d-%b-%y")

print(df.dtypes)

print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["Amount"], bins=30, kde=True)
plt.title("Satış məbləğinin paylanması")
plt.xlabel("Amount ($)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x="Country", y="Amount", data=df)
plt.xticks(rotation=45)
plt.title("Ölkələr üzrə satış məbləği")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\User\Desktop\Chocolate Sales.csv")

df["Amount"] = df["Amount"].replace(r"[\$,]", "", regex=True).astype(float)

corr_matrix = df.select_dtypes(include=["number"]).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasiya Matrisinin İstilik Xəritəsi")
plt.show()
```
